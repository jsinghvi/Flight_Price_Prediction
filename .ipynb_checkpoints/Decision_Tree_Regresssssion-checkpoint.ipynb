```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Flight_Ticket_Participant_Datasets/Data_Train.xlsx")
data.head()
```

```python
data = data.drop(["Additional_Info"],axis=1)
data.head()
```

```python
label = data["Price"]
# label.head()
data = data.drop(["Route","Price"],axis=1)
data.head()
```

```python
temp_date = data["Date_of_Journey"]
temp_date.head()
```

```python
from datetime import date
no_of_days = []
weekday = []

for i in temp_date:
    s = i.split("/")
    year = int(s[2])
    month = int(s[1])
    day = int(s[0])
    
    x = date(year,month,day)
    y = date.today()
    
    weekday.append(x.weekday())
    
    no_of_days.append((x-y).days)
```

```python
data["No_of_days"] = no_of_days
data["Weekday"] = weekday
data.head()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()


le.fit(data["Airline"])
data["Airline"] = le.transform(data["Airline"])

le.fit(data["Source"])
data["Source"] = le.transform(data["Source"])

le.fit(data["Destination"])
data["Destination"] = le.transform(data["Destination"])

data.head()
```

```python
day = data["Weekday"]
expensive_ticket_day = []

for d in day:
    if d==4 or d==5:
        expensive_ticket_day.append(1)
    else:
        expensive_ticket_day.append(0)
        
data["Expensive_Day"] = expensive_ticket_day
```

```python
arrival_time = data["Arrival_Time"]
temp = []

for i in arrival_time:
    temp.append(i[:5])
    
data = data.drop(["Arrival_Time"],axis=1)
data["Arrival_Time"] = temp
data.head()
```

```python
overnight_flight = []
dep = data["Dep_Time"]
arr = data["Arrival_Time"]
for i,j in zip(dep,arr):
    if i>j:
        overnight_flight.append(1)
    else:
        overnight_flight.append(0)

data["Overnight_Flight"] = overnight_flight
```

```python
data = data.drop(["Date_of_Journey"],axis=1)
data.head()
```

```python
import re
new_dur = []
duration = data["Duration"]
overnight = data["Overnight_Flight"]

for string in duration:
    s = str(string)
    a = []
    a = re.findall(r'\d+', s)
    if len(a)==2:
        new_dur.append((int(a[0])*60)+int(a[1]))
    else:
        new_dur.append(int(a[0])*60)

new_dur

# for i,j in zip(duration,overnight):
#     if j==1:
#         hours,minutes = i.split(" ")
# #         hours = hours[:-1]
#         minutes = minutes[:-1]
#         new_dur.append((hours*60)+minutes)
#     else:
#         hours = i.split(" ")
#         new_dur.append((hours[:-1])*60)
# new_dur
```

```python
stops = data["Total_Stops"]
new_Stops = []
for i in stops:
    if i=="non-stop":
        new_Stops.append(0)
    elif i=="1 stop":
        new_Stops.append(1)
    elif i=="2 stops":
        new_Stops.append(2)
    else:
        new_Stops.append(3)

data = data.drop(["Dep_Time","Arrival_Time","Duration","Total_Stops"],axis=1)
data["Duration"] = new_dur
data["Total_Stops"] = new_Stops
```

```python
data = data.drop(["Expensive_Day"],axis=1)
data.head()
```

```python
# data.to_csv("training_Data.csv")
# label.to_csv("training_labels.csv")
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(data,label)
```
